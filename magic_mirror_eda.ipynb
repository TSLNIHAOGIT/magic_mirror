```python
root_path='./data'
```

```python
import os
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv(os.path.join(root_path,'train.csv'),parse_dates=['auditing_date', 'due_date', 'repay_date'],
                    dtype={'user_id':np.uint32, 'listing_id':np.uint32, 
                                #'due_amt':np.float32,'repay_amt':np.float32
                               })
```

```python
train.sample(10)
```

```python
def replace_nan_date(x):
    if x=='\\N':
        return '1000-01-01'
    else:
        return x
def replace_nan_amt(x):
    if x=='\\N':
        return 0.0
    else:
        return x
train['repay_date']=train['repay_date'].map(replace_nan_date)
train['repay_amt']=train['repay_amt'].map(replace_nan_amt)
```

```python
train.sample(10)
```

```python
train.info(memory_usage='deep')
```

```python
def replace_nan_date(x):
    if x=='1000-01-01':
        return '2080-01-01'
    else:
        return x

train['repay_date']=train['repay_date'].map(replace_nan_date)

```

```python
#convert to datatime dtype
train['auditing_date'] = pd.to_datetime(train['auditing_date'], format ='%Y-%m-%d')
train['due_date'] = pd.to_datetime(train['due_date'], format ='%Y-%m-%d')
train['repay_date'] = pd.to_datetime(train['repay_date'], format ='%Y-%m-%d')
#convert to unint32
for each in ['user_id','listing_id']:
    train[each]=train[each].astype(np.uint32)
#convert to float32
for each in ['due_amt','repay_amt']:
    train[each]=train[each].astype(np.float32)
```

```python
train.info(memory_usage='deep')
```

```python
train.sample(10)
```

```python
for each_dtype in ['int','float','']:
    mean_usage_b=train.select_dtypes(include=[each_dtype]).memory_usage(deep=True).mean()
    mean_usage_mb = mean_usage_b /1024**2

    print("Average memory usage for {} columns: {:03.2f} MB".format(each_dtype,mean_usage_mb))
```
